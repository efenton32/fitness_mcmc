```python
#Standard Imports
import numpy as np
import pymc3 as pm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az
```

```python
#Imports from fitness mcmc package 
import fitness_mcmc
import fitness_mcmc.data_io as io 
import fitness_mcmc.fitness_mcmc as m
```

```python
"""
Loading data and looking at the first 40 ordered genotypes. 
"""
data, time, ordered_frequencies  = io.load_data('filtered_counts_ypd_temp_30.txt')
data
```

```python
fitness_model = m.Fitness_Model(ordered_frequencies[0:400],time)
fitness_model.mcmc_sample()
```

```python
fitness_model.plot_mcmc_trace()
```

```python
fitness_model.find_MAP()
fitness_model.plot_MAP_estimate(type = "lin")
```

```python
s = np.linspace(-0.25,0.25,1000)
sigma = fitness_model.map_estimate["sigma"]
mu = fitness_model.map_estimate["mu"]
s_i = 1/(np.sqrt(np.pi*2)*sigma)*np.exp(-(s-mu)**2/(2*sigma**2))
plt.plot(s,s_i)

s = fitness_model.map_estimate["s"]
normal = -np.trapz(np.squeeze(s))
print(normal)
bins = np.linspace(-0.25,0,70)
plt.hist(s,bins = bins,density = True)
plt.show()
```
